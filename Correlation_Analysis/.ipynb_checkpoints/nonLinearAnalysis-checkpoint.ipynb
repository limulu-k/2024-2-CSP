```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import ccf, grangercausalitytests, coint
import seaborn as sns
import os
```

```python
# Load the datasets
climate_data_path = '../ClimateDataTeam/climate_data/merged_weekly_avg_temp.csv'
air_pollutants_path = '../NewData/Weekly_Air_Pollutants.csv'
greenhouse_gas_path = '../NewData/Weekly_Greenhouse_Gas.csv'
population_path = '../NewData/Weekly_Population.csv'
power_usage_path = '../NewData/Weekly_Power_Usage.csv'

main_target = air_pollutants_path
```

```python
# 데이터 불러오기
main_target = pd.read_csv(air_pollutants_path)
climate_data_df = pd.read_csv(climate_data_path)

# datetime 컬럼을 datetime 객체로 변환
main_target['datetime'] = pd.to_datetime(main_target['datetime'])
climate_data_df['datetime'] = pd.to_datetime(climate_data_df['datetime'])

# 데이터 병합
merged_df = pd.merge(main_target, climate_data_df, on='datetime', how='inner')
```

```python
# 수치형 변수만 선택
air_pollutants_cols = main_target.select_dtypes(include=[np.number]).columns
climate_data_cols = climate_data_df.select_dtypes(include=[np.number]).columns

# 분석 결과 저장용
cross_corr_results = []
granger_results = []
coint_results = []
```

```python
# 기후 데이터 ↔ 오염 데이터 쌍만 분석
for var1 in air_pollutants_cols:
    for var2 in climate_data_cols:
        # 교차상관 분석 (CCF)
        cross_corr = ccf(merged_df[var1], merged_df[var2])[:20]  # 최대 20 lag
        max_cross_corr = max(cross_corr)
        cross_corr_results.append((var1, var2, max_cross_corr))
        
        # Granger 인과관계 분석
        try:
            granger_test = grangercausalitytests(merged_df[[var1, var2]], maxlag=5, verbose=False)
            min_p_value = min(granger_test[lag][0]['ssr_ftest'][1] for lag in granger_test.keys())
            granger_results.append((var1, var2, min_p_value))
        except Exception as e:
            granger_results.append((var1, var2, None))  # 실패 시 None

        # 공적분 분석
        try:
            coint_score, coint_p_value, _ = coint(merged_df[var1], merged_df[var2])
            coint_results.append((var1, var2, coint_p_value))
        except Exception as e:
            coint_results.append((var1, var2, None))  # 실패 시 None
```

```python
# 교차상관 결과 출력
best_cross_corr = max(cross_corr_results, key=lambda x: abs(x[2]))
print(f"가장 높은 교차상관: 변수1={best_cross_corr[0]}, 변수2={best_cross_corr[1]}, 상관계수={best_cross_corr[2]:.4f}")

# Granger 인과관계 결과 출력
best_granger = min(granger_results, key=lambda x: x[2] if x[2] is not None else float('inf'))
print(f"가장 낮은 Granger 인과관계 p-value: 변수1={best_granger[0]}, 변수2={best_granger[1]}, p-value={best_granger[2]:.4f}")

# 공적분 분석 결과 출력
best_coint = min(coint_results, key=lambda x: x[2] if x[2] is not None else float('inf'))
print(f"가장 낮은 공적분 p-value: 변수1={best_coint[0]}, 변수2={best_coint[1]}, p-value={best_coint[2]:.4f}")
```

```python
# 교차상관 히트맵 생성
cross_corr_matrix = pd.DataFrame(index=air_pollutants_cols, columns=climate_data_cols)
for var1, var2, corr in cross_corr_results:
    cross_corr_matrix.at[var1, var2] = corr
cross_corr_matrix = cross_corr_matrix.astype(float)

plt.figure(figsize=(12, 8))
sns.heatmap(cross_corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Cross-Correlation Matrix (Air Pollutants vs Climate Data)")
plt.show()
```

```python
# Granger 인과관계 히트맵 생성
granger_matrix = pd.DataFrame(index=air_pollutants_cols, columns=climate_data_cols)
for var1, var2, p_value in granger_results:
    granger_matrix.at[var1, var2] = p_value
granger_matrix = granger_matrix.astype(float)

plt.figure(figsize=(12, 8))
sns.heatmap(granger_matrix, annot=True, fmt=".4f", cmap="coolwarm")
plt.title("Granger Causality p-value Matrix (Air Pollutants vs Climate Data)")
plt.show()
```

```python
# 공적분 히트맵 생성
coint_matrix = pd.DataFrame(index=air_pollutants_cols, columns=climate_data_cols)
for var1, var2, p_value in coint_results:
    coint_matrix.at[var1, var2] = p_value
coint_matrix = coint_matrix.astype(float)

plt.figure(figsize=(12, 8))
sns.heatmap(coint_matrix, annot=True, fmt=".4f", cmap="coolwarm")
plt.title("Cointegration p-value Matrix (Air Pollutants vs Climate Data)")
plt.show()
```
