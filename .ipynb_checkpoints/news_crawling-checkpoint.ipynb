```python
def downloadData(code, start, end):
    try:
        service = Service(ChromeDriverManager().install())
        driver = webdriver.Chrom(service=service)
        l = len(start)
        res = []
        
        for idx in range(int(start), int(end)+1):
            try:
                url = f"https://n.news.naver.com/mnews/article/{code}/{idx:0>{l}}"
                driver.get(url)
                time.sleep(0.8)
                if "#dic_area" in driver.page_source:
                    title = driver.find_element(By.CSS_SELECTOR, "#title_area").text
                    p = driver.find_element(By.CSS_SELECTOR, "#dic_area").text
                    res.append([url,p])
            except:
                print(f"{url} is ERROR")
        driver.quit()
        return res
```

```python
def find_start(code, e):
    start = 0
    end = int(e)
    l = len(e)
    print(f"start:{start}\tend:{end}")
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service)
    
    while start < end-1:
        mid = (start+end)//2
        try:
            url = f"https://n.news.naver.com/mnews/article/{code}/{mid:0>{l}}"
            print(f"start:{start}\tend:{end} => mid{mid}")
            print(url)
            driver.get(url)
            time.sleep(0.8)
            if "언론사 요청에 의해 삭제된 기사입니다." in driver.page_source or "#dic_area" in driver.page_source:
                end = mid
            else:
                start = mid
        except:
            start = mid
    
    start += 1
    index = 100000

    while index > 0:
        url = f"https://n.news.naver.com/mnews/article/{code}/{start:0>{l}}"
        print(f"finding recent news\nurl:{url}")
        driver.get(url)
        time.sleep(0.8)
        if "언론사 요청에 의해 삭제된 기사입니다." in driver.page_source or "페이지를 찾을 수 없습니다" in driver.page_source:
            start += index
        else:
            start -= index
            index = index // 2
    driver.quit()
    res = f"{start+1:0>{l}}"
    print(f"Result : {res}")
    return [code, res]
```

```python
# 쓰기 작업을 보호하기 위한 Lock 객체 생성
file_lock = threading.Lock()

# 스레드에서 실행할 함수
def write_to_file(file_name, text):
    with file_lock:  # 파일 쓰기 작업에 Lock을 적용
        with open(file_name, 'a') as f:  # 파일을 append 모드로 엽니다.
            for line in text:
                f.write(line + '\n')
                print(f"{line} 쓰기 완료")
                # 파일에 쓰는 동안 다른 스레드가 접근하지 못하도록 보호됨

# 쓰기할 데이터
data1 = ["스레드1 - 데이터1", "스레드1 - 데이터2", "스레드1 - 데이터3"]
data2 = ["스레드2 - 데이터1", "스레드2 - 데이터2", "스레드2 - 데이터3"]

# 스레드 생성
thread1 = threading.Thread(target=write_to_file, args=("output.txt", data1))
thread2 = threading.Thread(target=write_to_file, args=("output.txt", data2))

# 스레드 시작
thread1.start()
thread2.start()

# 메인 스레드에서 두 스레드가 끝날 때까지 대기
thread1.join()
thread2.join()

print("모든 파일 쓰기 작업 완료")
```

```python
# 메인 함수
# 신문사 코드:[신문사 이름. 최근 뉴스] 로 크롤링할 데이터셋 지정
typeOfNewsCompany = {"028":["한겨례","0002710844"]}
# 매일경제 : https://www.mk.co.kr/news/society/100

file_lock = threading.Lock()

results = []

# ThreadPoolExecutor 사용
with concurrent.futures.ThreadPoolExecutor(max_workers=3) as executor:
    futures = [executor.submit(find_start, key, value[1]) for key, value in typeOfNewsCompany.items()]
    
    # 완료된 작업의 결과를 결과 리스트에 추가
    for future in concurrent.futures.as_completed(futures):
        print(future)
        tmp = future.result()
        typeOfNewsCompany[tmp[0]] = [typeOfNewsCompany[tmp[0]][0],tmp[1],typeOfNewsCompany[tmp[0]][1]]

print("Results:", results)
```
