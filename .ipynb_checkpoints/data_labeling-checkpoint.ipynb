```python
import pandas as pd
import os
import warnings

# ê²½ê³  ë©”ì‹œì§€ ë¬´ì‹œ
warnings.filterwarnings("ignore", category=UserWarning, module="openpyxl")

# CSV íŒŒì¼ ìƒì„± í•¨ìˆ˜ (í—¤ë” í¬í•¨)
def create_csv_file(file_path):
    # ë¹ˆ ë°ì´í„°í”„ë ˆì„ ìƒì„±
    df = pd.DataFrame(columns=["content", "keyword", "label"])
    # íŒŒì¼ ì €ì¥
    df.to_csv(file_path, index=False, encoding="utf-8-sig")
    print(f"{file_path} íŒŒì¼ì´ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤.")

# í•œ ì¤„ì”© ë°ì´í„°ë¥¼ ì¶”ê°€í•˜ëŠ” í•¨ìˆ˜
def add_row_to_csv(file_path, content, keyword, label=None):
    # ìƒˆ ë°ì´í„° ìƒì„± (labelì€ None ê°’ìœ¼ë¡œ ë‘ )
    new_data = pd.DataFrame([[content, keyword, label]], columns=["content", "keyword", "label"])
    
    # íŒŒì¼ì´ ì¡´ì¬í•˜ë©´ í—¤ë” ì—†ì´ ì¶”ê°€
    if os.path.exists(file_path):
        new_data.to_csv(file_path, mode='a', header=False, index=False, encoding="utf-8-sig")
    else:
        print("íŒŒì¼ì´ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤. ë¨¼ì € íŒŒì¼ì„ ìƒì„±í•´ì£¼ì„¸ìš”.")

def clear_content(content):
    # 2ê°œ ì´ìƒì˜ ì—°ì†ëœ ê°œí–‰(\n)ì„ 1ê°œë¡œ ë³€ê²½
    cleaned_content = re.sub(r'\n{2,}', '\n', content)
    return cleaned_content

# CSV íŒŒì¼ ì œê±° í›„ ìƒì„±
file_path = f"{os.getcwd()}\\newsData\\label_data.csv"
if os.path.exists(file_path):
    os.remove(file_path)
    print(f"{file_path} íŒŒì¼ì´ ì‚­ì œë˜ì—ˆìŠµë‹ˆë‹¤.")
else:
    print("íŒŒì¼ì´ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")
create_csv_file(file_path)

# ê¸°ì¡´ ì—‘ì…€ ë°ì´í„° ì½ê¸°
xlsxPath = os.path.join(dataPath, "datas")
xlsxList = os.listdir(xlsxPath)
print(f"ì—‘ì…€ ê²½ë¡œ : {xlsxPath}")


for x in range(10):
    tmp = os.path.join(xlsxPath, xlsxList[x])
    print(tmp)
    df = pd.read_excel(tmp)
    idx = 0
    
    for row in df.itertuples(index=False):
        keyWord = list(map(str, row[15].split(",")))
        content = row[16]
        content = clear_content(content)
        add_row_to_csv(file_path, content, keyWord)
        if idx == 99:
            break
        idx += 1
print("Making labeling file complete")
```

```python
import os
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments

# ë°ì´í„° ê²½ë¡œ ì„¤ì •
dataPath = os.path.abspath("./newsData")
labeled_data_path = os.path.join(dataPath, "labeled_data.csv")

# ë°ì´í„° ë¡œë“œ
labeled_df = pd.read_csv(labeled_data_path)

# ë°ì´í„°ì…‹ ì¤€ë¹„
texts = (labeled_df['keyword'] + " " + labeled_df['content']).tolist()  # keywordì™€ content ê²°í•©
labels = labeled_df['label'].astype(int).tolist()
print(labels)

# ë°ì´í„° ë¶„í•  (í›ˆë ¨/ê²€ì¦)
train_texts, val_texts, train_labels, val_labels = train_test_split(texts, labels, test_size=0.2, random_state=42)

# KoBERT í† í¬ë‚˜ì´ì € ë° ëª¨ë¸ ë¡œë“œ
tokenizer = AutoTokenizer.from_pretrained("skt/kobert-base-v1")
model = AutoModelForSequenceClassification.from_pretrained("skt/kobert-base-v1", num_labels=7)  # 7ê°œ ê°ì • ë¶„ë¥˜

# í…ìŠ¤íŠ¸ë¥¼ í† í°í™”
train_encodings = tokenizer(train_texts, padding=True, truncation=True, return_tensors="pt", max_length=64)
val_encodings = tokenizer(val_texts, padding=True, truncation=True, return_tensors="pt", max_length=64)

# ë°ì´í„°ì…‹ í´ë˜ìŠ¤ ì •ì˜
class NewsDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: val[idx] for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

# ë°ì´í„°ì…‹ ì¤€ë¹„
train_dataset = NewsDataset(train_encodings, train_labels)
val_dataset = NewsDataset(val_encodings, val_labels)

# í›ˆë ¨ ì¸ì ì„¤ì •
training_args = TrainingArguments(
    output_dir=output_dir,
    num_train_epochs=3,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    warmup_steps=500,
    weight_decay=0.01,
    logging_steps=500,
    evaluation_strategy="epoch",
    save_strategy="epoch"
)

# Trainer ê°ì²´ ìƒì„±
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    data_collator=lambda data: {'input_ids': torch.stack([f['input_ids'] for f in data]),
                            'attention_mask': torch.stack([f['attention_mask'] for f in data]),
                            'labels': torch.tensor([f['labels'] for f in data])}
)

# ëª¨ë¸ í›ˆë ¨
trainer.train()

# ëª¨ë¸ ì €ì¥
model.save_pretrained(os.path.join(dataPath, "finetuned_kobert"))
tokenizer.save_pretrained(os.path.join(dataPath, "finetuned_kobert"))
```

```python
import os
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import warnings

# ê²½ê³  ë¬´ì‹œ
warnings.filterwarnings("ignore", category=UserWarning, module='openpyxl')
os.environ["HF_HUB_DISABLE_SYMLINKS_WARNING"] = "1"

# ê²½ë¡œ ì„¤ì •
datasPath = os.path.join(dataPath, "datas")
xlsxList = os.listdir(datasPath)
print(xlsxList)

# # í›ˆë ¨ëœ ëª¨ë¸ê³¼ í† í¬ë‚˜ì´ì € ë¡œë“œ
model_path = os.path.join(dataPath, "finetuned_kobert")  # í›ˆë ¨ëœ ëª¨ë¸ì˜ ê²½ë¡œë¡œ ë³€ê²½
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSequenceClassification.from_pretrained(model_path)

# ê°ì • ë¶„ì„ í•¨ìˆ˜ ì •ì˜
def analyze_sentiment(keywords, content):
    # í‚¤ì›Œë“œì™€ ë³¸ë¬¸ ê²°í•©
    text = f"{content} í‚¤ì›Œë“œ: {', '.join(keywords)}"
    
    # ì…ë ¥ í…ìŠ¤íŠ¸ë¥¼ í† í°í™” - token_type_idsë¥¼ ìƒì„±í•˜ì§€ ì•Šë„ë¡ ì„¤ì •
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512, return_token_type_ids=False)
    
    # ëª¨ë¸ì— ì…ë ¥í•˜ì—¬ ì˜ˆì¸¡ ìˆ˜í–‰
    with torch.no_grad():
        outputs = model(**inputs)
        predictions = torch.argmax(outputs.logits, dim=-1)
    
    # ì˜ˆì¸¡ ê²°ê³¼ í•´ì„ (0: ê¸ì •, 1: ë¶€ì •ìœ¼ë¡œ ê°€ì •)
    return "positive" if predictions.item() == 0 else "negative"


# ê°ì • ë¶„ì„ í•¨ìˆ˜ ì •ì˜
# def analyze_sentiment(keywords, content):
#     # í‚¤ì›Œë“œì™€ ë³¸ë¬¸ ê²°í•©
#     text = f"{content} í‚¤ì›Œë“œ: {', '.join(keywords)}"
    
#     # ì…ë ¥ í…ìŠ¤íŠ¸ë¥¼ í† í°í™”
#     inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    
#     # ëª¨ë¸ì— ì…ë ¥í•˜ì—¬ ì˜ˆì¸¡ ìˆ˜í–‰
#     with torch.no_grad():
#         outputs = model(**inputs)
#         predictions = torch.argmax(outputs.logits, dim=-1)
    
#     # ì˜ˆì¸¡ ê²°ê³¼ í•´ì„ (0: ê¸ì •, 1: ë¶€ì •ìœ¼ë¡œ ê°€ì •)
#     return "positive" if predictions.item() == 0 else "negative"

# ë°ì´í„° íŒŒì¼ ì½ê¸° ë° ë¶„ì„ ìˆ˜í–‰
for xl in xlsxList:
    tmp = os.path.join(datasPath, xl)
    print(tmp)
    print(f"Processing file: {tmp}")
    df = pd.read_excel(tmp)
    
    idx = 0
    for row in df.itertuples(index=False):
        # í‚¤ì›Œë“œì™€ ë³¸ë¬¸ ì¶”ì¶œ
        keywords = list(map(str, row[15].split(",")))
        content = row[16]
        
        # ê°ì • ë¶„ì„ ìˆ˜í–‰
        sentiment_result = analyze_sentiment(keywords, content)
        print(f"ì œëª©: {row[4]}\tê²°ê´ê°’ : {sentiment_result}")
        
        # ì„ì‹œë¡œ 40ê°œ ë°ì´í„°ë§Œ ì²˜ë¦¬ (í…ŒìŠ¤íŠ¸ìš©)
        if idx == 40:
            break
        idx += 1
```
